```python
# Import necessary libraries
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Prepare the data
data = [[10, 30], [40, 14], [33, 79], [67, 21], [10, 99], [78, 56]]

# Create and fit the KMeans model
kmeans = KMeans(n_clusters=2, random_state=40)
kmeans.fit(data)
# Get the labels and cluster centers
labels = kmeans.labels_
centers = kmeans.cluster_centers_
# Print the results
print("Labels:", labels)
print("Cluster Centers:", centers)

# Optional: Visualize the clusters
plt.scatter([point[0] for point in data], [point[1] for point in data], c=labels, cmap='viridis', marker='o')
plt.scatter(centers[:, 0], centers[:, 1], c='red', marker='X', s=200)  # Cluster centers
plt.title('K-Means Clustering')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid()
plt.show()
```
