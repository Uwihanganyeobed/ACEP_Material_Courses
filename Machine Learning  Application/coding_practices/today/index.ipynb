```python
# cluster customers into different groups based on their purchase history and spending score
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import numpy as np

#load data set
data_set = {
    'purchase_history':[100,200,300,400,500,600,700,800,900,1000],
    'spending_score':[15,24,74,41,50,105,65,81,33,99]
}

dataFrame = pd.DataFrame(data_set)
dataFrame

#apply kmeans clustering

kmeans = KMeans(n_clusters=5, random_state=42)
dataFrame['Cluster'] = kmeans.fit_predict(dataFrame)

#visualize clusters
plt.scatter(dataFrame['purchase_history'],dataFrame['spending_score'],c=dataFrame['Cluster'],cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=300,c='red',label='Centroids')
plt.xlabel('Purchase History')
plt.ylabel('Spending Score')
plt.title('Customer Clusters')
plt.legend()
plt.show()



```
